```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Read the original image
image1 = '/content/drive/MyDrive/2.jpeg'
img = cv2.imread(image1,flags=0)
# Blur the image for better edge detection
img_blur = cv2.GaussianBlur(img,(3,3), sigmaX=0, sigmaY=0)
cv2_imshow(img)
print('Original Image\n')
cv2_imshow(img_blur)
print('Blurred Image')
```

```python
#canny edge detection is done on the gradient direction itself
edges = cv2.Canny(image= img_blur, threshold1=100, threshold2=200)
#this uses the hysterisis method of finding whether any pixel is an edge or not
cv2_imshow(edges)
print('edges detected using Canny Edge Detector')
```
