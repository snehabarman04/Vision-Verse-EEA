```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
image1 = '/content/drive/MyDrive/road.png'
image2 = '/content/drive/MyDrive/3.jpeg'
image3 = '/content/drive/MyDrive/4.jpg'

#reading the image using the cv2 function
img = cv2.imread(image3,-1)
cv2_imshow(img)
print('Original Image\n')

#blurring out the image a little bit first
img_blur = cv2.GaussianBlur(img,(3,3), sigmaX=0, sigmaY=0)
cv2_imshow(img_blur)
print('Blurred Image\n')

#finding out the edges using the Canny edge detector
edges = cv2.Canny(image= img_blur, threshold1=100, threshold2=200)
cv2_imshow(edges)
print('Edges of the Image using Canny Edge Detector')

```

```python
#performing the hough transform
lines = cv2.HoughLinesP(edges, 1, np.pi/180, threshold=10, minLineLength=10, maxLineGap=200)
for line in lines:
  x1,y1,x2,y2 = line[0]
  cv2.line(img,(x1,y1),(x2,y2),(0, 0, 255), 1)

cv2_imshow(img)
```
