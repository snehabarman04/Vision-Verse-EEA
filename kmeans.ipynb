```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
img1='/content/drive/MyDrive/1.png'
img2='/content/drive/MyDrive/2.png'
img3='/content/drive/MyDrive/3.png'
img4='/content/drive/MyDrive/4.png'
img5='/content/drive/MyDrive/5.png'
img6='/content/drive/MyDrive/6.jpg'
```

```python
image1=cv2.imread(img1)
image2=cv2.imread(img2)
image3=cv2.imread(img3)
image4=cv2.imread(img4)
image5=cv2.imread(img5)
image6=cv2.imread(img6)
```

```python
pixels = image1.reshape((-1, 3))


plt.figure(figsize=(15,5))
# Plot original image
plt.subplot(4, 4, 2)
plt.imshow(image1)
plt.title('Original Image 1')


criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
_, labels, centers = cv2.kmeans(np.float32(pixels), 2, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
#ignoring the compactness parameter out of the three values that K-means returns


# Reshaping the labels to match the original image shape
segmented_image1 = labels.reshape(image1.shape[:2])
plt.subplot(4, 4, 9)
plt.imshow(segmented_image1)  # Use a colormap suitable for discrete values
plt.title('K-Means clustering done with K=2')


criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
_, labels, centers = cv2.kmeans(np.float32(pixels), 4, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
#ignoring the compactness parameter out of the three values that K-means returns


# Reshaping the labels to match the original image shape
segmented_image1 = labels.reshape(image1.shape[:2])
plt.subplot(4, 4, 10)
plt.imshow(segmented_image1)  # Use a colormap suitable for discrete values
plt.title('K-Means clustering done with K=4')


criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
_, labels, centers = cv2.kmeans(np.float32(pixels), 6, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
#ignoring the compactness parameter out of the three values that K-means returns


# Reshaping the labels to match the original image shape
segmented_image1 = labels.reshape(image1.shape[:2])
plt.subplot(4, 4, 11)
plt.imshow(segmented_image1)  # Use a colormap suitable for discrete values
plt.title('K-Means clustering done with K=6')
```
