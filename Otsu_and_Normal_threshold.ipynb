```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
img='/content/drive/MyDrive/boat.jpg'
original= cv2.imread(img)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(original)
plt.title('Original Image')
```

```python
plt.subplot(2, 3, 2)
plt.imshow(original)
plt.title('Original Image')


# performing Otsu's thresholding

#first converting the image to grayscale
gray_image = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)
_, otsu_threshold = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.subplot(2, 3, 4)
plt.imshow(otsu_threshold)
plt.title('Otsu\'s Thresholding')


# Normal thresholding
_, normal_threshold = cv2.threshold(gray_image, 108, 100, cv2.THRESH_BINARY)

plt.subplot(2, 3, 6)
plt.imshow(normal_threshold)
plt.title('Normal Thresholding')
```
